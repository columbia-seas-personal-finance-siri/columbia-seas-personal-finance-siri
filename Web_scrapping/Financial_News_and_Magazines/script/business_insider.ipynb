```python
import requests
from bs4 import BeautifulSoup
import json
import time
import datetime
d_l=[]
URL = "https://www.businessinsider.com/clusterstock"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
results=soup.find("section",class_="top-vertical-trio wrapper row")
results=results.find_all("a",class_="tout-title-link")
results=[URL[0:-13]+i.get("href") for i in results]
for i in results:
  page = requests.get(i)
  soup1 = BeautifulSoup(page.content, "html.parser")
  title=soup1.find("title").text
  date=soup1.find("meta",{"name":"dateModified"}).get("content")
  date=datetime.datetime.strptime(date, '%Y-%m-%dT%H:%M:%S%z')
  label=soup1.find("meta",{"name":"news_keywords"}).get("content")
  content=soup1.find("div",class_="content-lock-content")
  content=content.find_all("p")
  content="".join([i.text for i in content])
  d={"title":title,"date":date,"content":content,"label":label,"link":i}
  d_l.append(d)
```

```python
results=soup.find_all("section",class_="river-item featured-post")
for i in results:
  #print(i.prettify())
  r=i.find("a",class_="tout-title-link").get("href")
  page = requests.get(URL[0:-13]+r)
  soup1 = BeautifulSoup(page.content, "html.parser")
  title=soup1.find("title").text
  date=soup1.find("meta",{"name":"dateModified"}).get("content")
  date=datetime.datetime.strptime(date, '%Y-%m-%dT%H:%M:%S%z')
  label=soup1.find("meta",{"name":"news_keywords"}).get("content")
  content=soup1.find("div",class_="content-lock-content")
  content=content.find_all("p")
  content="".join([i.text for i in content])
  d={"title":title,"date":date,"content":content,"label":label,"link":URL[0:-13]+r}
  d_l.append(d)
```

```python
d_json=json.dumps(d_l,default=str)
d_json
```

```python
with open("business_insider.json", "w") as outfile:
    outfile.write(d_json)
```
