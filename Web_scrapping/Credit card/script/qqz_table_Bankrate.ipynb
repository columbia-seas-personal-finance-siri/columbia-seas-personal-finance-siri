```python
from bs4 import BeautifulSoup
import pandas as pd
def fetch_page(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.text
    except requests.RequestException as e:
        print(f"An error occurred: {e}")
        return None
def create_credit_card_dataframe(html_content):
    # Initialize BeautifulSoup object
    soup = BeautifulSoup(html_content, 'html.parser')
    # Find the table
    table = soup.find('div', {'class': 'TwoColumnLayout TwoColumnLayout--reversed gap-y-0'})

    if table is None:
        print("Table not found")
        return None
    
    # Find all rows in the table body
    tbody = table.find('tbody')
    
    if tbody is None:
        print("Table body not found")
        return None
    
    rows = tbody.find_all('tr')
    
    # Initialize lists to hold scraped data
    card_names = []
    our_picks = []
    card_highlights = []
    bankrate_scores = []
    
    for row in rows:
        cells = row.find_all('td')
        
        if len(cells) == 4:
            card_names.append(cells[0].text.strip())
            our_picks.append(cells[1].text.strip())
            card_highlights.append(cells[2].text.strip())
            bankrate_scores.append(cells[3].text.strip().split('\n')[0])
    
    # Create a DataFrame
    df = pd.DataFrame({
        'Card Name': card_names,
        'Our Pick For': our_picks,
        'Card Highlights': card_highlights,
        'Bankrate Score': bankrate_scores
    })
    
    return df


url = "https://www.bankrate.com/finance/credit-cards/todays-best-credit-cards/"
html_content = fetch_page(url)

# If fetching was successful, proceed to extract card info
if html_content:
    # Extract card info and convert it to a Pandas DataFrame
    df = create_credit_card_dataframe(html_content)
    print(df)
print(df)
```
