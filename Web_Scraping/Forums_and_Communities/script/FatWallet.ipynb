```python
import requests
from bs4 import BeautifulSoup
import json
import time
import datetime
d_l=[]
URL = "https://www.fatwallet.com/forums/finance"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
results=soup.find("div",class_="css-1rdeovq")
label=results.find("h2",class_="css-8lq3ou").text
results=results.find_all("div",class_="css-1as73dl e15g0fbj0")
d_l=[]
for i in results:
    a=i.find("div",class_="css-h7sfyo")
    aa=i.find("a").get("href")
    b=a.find("div",class_="css-70qvj9").text
    c=a.find("span",class_="css-105ngdy").text
    d=a.find("span",class_="css-1cona4n").text
    d={"brand":b,"coupon":c,"detail":d,"label":label,"link":aa}
    d_l.append(d)
```

```python
import pandas as pd
pd.DataFrame(d_l)
```

```python
results=soup.find("div",class_="css-1yxmrz6")
label=results.find("h2",class_="css-1bwkg98").text
ii=results.find_all("div",class_="css-1as73dl e15g0fbj0")
for j in ii:
    #try:
    a=j.find("a").get("href")
    b=j.find("span",class_="css-142n6z8").text
    c=j.find("span",class_="css-1o3lf2p").text
    d=j.find("span",class_="css-1f5exmu").text
    d={"brand":b,"coupon":c,"detail":d,"label":label,"link":a}
    d_l.append(d)
            
            
```

```python
nx=results.find_next(class_="css-1yxmrz6")
label=nx.find("h2",class_="css-1bwkg98").text
coupons_l=nx.find_all("a")
for i in coupons_l:
    link=i.get("href")
    detail=i.find("span",class_="css-oqldcf").text
    coupon=i.find("span",class_="css-1o3lf2p").text
    brand=i.find("img").get("alt").replace(" logo","")
    d={"brand":brand,"coupon":coupon,"detail":detail,"label":label,"link":link}
    d_l.append(d) 
```
