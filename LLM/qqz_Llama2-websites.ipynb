#**extract the data from these URLs**

```python
URLs=[
    'https://www.morningstar.com/personal-finance/best-way-save-college',
    'https://www.morningstar.com/specials/what-is-a-529-college-savings-plan',
    'https://www.morningstar.com/personal-finance/themes-across-529-plan-landscape-6-charts'

]
```

```python
loaders=UnstructuredURLLoader(urls=URLs)
data=loaders.load()
```

```python
data
```

```python
len(data)
```

#**Split the Text into Chunks**

```python
text_splitter=CharacterTextSplitter(separator='\n',
                                    chunk_size=1000,
                                    chunk_overlap=200)
```

```python

text_chunks=text_splitter.split_documents(data)
```

#**Download the Hugging Face Embeddings**

```python
#embeddings=HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
embeddings=HuggingFaceEmbeddings()
```

```python
query_result = embeddings.embed_query("Hello world")
len(query_result)
```

#**Convert the Text Chunks into Embeddings and Create a Knowledge Base**

```python
vectorstore=FAISS.from_documents(text_chunks, embeddings)
```

#**Create a Large Language Model (LLM) Wrapper**

```python
llm
```

```python
notebook_login()
```

```python
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                          use_auth_token=True,)


model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                             device_map='auto',
                                             torch_dtype=torch.float16,
                                             use_auth_token=True,
                                              load_in_8bit=True,
                                              #load_in_4bit=True
                                             )
```

```python
pipe = pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 512,
                do_sample=True,
                top_k=30,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
```

```python
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0})
```

```python
llm.predict("Please provide a concise summary of 529 Plans")
```

#**Initialize the Retrieval QA with Source Chain**

```python
chain = RetrievalQAWithSourcesChain.from_llm(llm=llm, retriever=vectorstore.as_retriever())
```

```python
result=chain({"question": "How good is 529 Plan?"}, return_only_outputs=True)
```
